## Create a Test Set

It is necessary to have training and testing dataset for a machine learning study. The accuracy of the algorithm trained with the training data will be tested for accuracy using a test set. If you identify any pattern in the test set, there is a high probability to select a algorithm which works well with the properties of the testing set. To avoid this situation, it is important to create a testing set at before the analysis and not having look at that until the training process finished.

The train_test_split method of sklearn is used in this study to divide data to training and testing data. Training to testing ratio is kept as 0.8: 0.2 which is commonly used ratio in machine learning studies. Setting random state to 42 will select same set of observations to training and testing data.

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data, test_size=0.2,
random_state=42)
```

According to the knowledge obtained from real estate company the median income of the district is highly important to predict housing prices. Therefore, it is important to select representative samples of each strata of the median income of the data set. 

Since the median income is a continuous numerical attribute, the dataset is divided to categories(strata) after careful observation of the dataset. The pd.cut() function is used in this study to categorize the median income attribute into five categories. The resulting dataset can be used to do stratified sampling based on the newly created income category. 

```python
data["income_cat"] = pd.cut(data["median_income"],
bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
labels=[1, 2, 3, 4, 5])
```

```python
# visualize income categories in a histogram
data["income_cat"].hist()
```

Now perform stratified sampling based on the income categories using scikit learn Stratified Shuffle Split tools. The for loop of the following code will split data into test and training sets according to the assigned ratio. The major advantage of using stratified sampling is it takes samples for testing and training from the sample strata already assigned by the user. 

```python
# import function
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42) # split data train:test to 8:2 ratio 
for train_index, test_index in split.split(data, data["income_cat"]):
    strat_train_set = data.loc[train_index]
    strat_test_set = data.loc[test_index]
```

Now the data is divided into training and testing sets. It is important to double check whether this process is done correctly as both sets should be representative of all income categories. Let's check the proportions of each income category in the testing data set. The same can be applied to the training data set. 

```python
#  Income category proportions in the test set
strat_test_set["income_cat"].value_counts() / len(strat_test_set)
```

Now it is clear that the testing data set is composed with all the income categories. Lets check how these proportions compare with the full dataset. The proportions should be similar in both cases. 

```python
#  Income category proportions in the test set
data["income_cat"].value_counts() / len(data)
```

It is obvious that the proportions are similar in each dataset. Which means that now we are good to proceed with the training adn testing data sets. Also, we can drop the income category attribute and return to the original data format.

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
```

## Transformation Pipelines

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6
```

```python
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self # nothing else to do
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
            bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```

```python
num_pipeline = Pipeline([
('imputer', SimpleImputer(strategy="median")),
('attribs_adder', CombinedAttributesAdder()),
('std_scaler', StandardScaler()),
])
housing_num_tr = num_pipeline.fit_transform(housing_num)
```

Feature scaling is an important transformation that should be done during a machine learning study. Because in most of the cases, Machine Learning algorithms do not perform well with numerical data in different scales.

In this case, number of rooms range from 6 to 39,320 and median income range from 0 to 15. This is two different scales.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])
housing_num_tr = num_pipeline.fit_transform(housing_num)
```

```python
from sklearn.compose import ColumnTransformer
num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])
housing_prepared = full_pipeline.fit_transform(housing)
```

## Select and Train a Model

To this point, we have explored data, created training and testing data and cleaned up the data using transformation pipelines. Now it is the time to select and train a machine learning algorithm.

let's train a Linear Regression model.

```python
from sklearn.linear_model import LinearRegression
# create the linear regression model
lin_reg = LinearRegression()
# fit the model to training features and labels
lin_reg.fit(housing_prepared, housing_labels)
```

Lets use the Linear Regression model with some instances of the training set.

```python
some_data = housing.iloc[:5]
some_labels = housing_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)
print("Predictions:", lin_reg.predict(some_data_prepared))

print("Labels:", list(some_labels))
```

The model can be used for predictions and looking at the predicted results and the labels, we can say that the predictions are not that accurate.
Let's measure the accuracy of the model using RMSE.

```python
from sklearn.metrics import mean_squared_error
housing_predictions = lin_reg.predict(housing_prepared)
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

Compared to the median house values range between 120,000 and 265,000, the prediction error of 68,628 is not enough. This shows that the model do not predict accurate enough, which happens during underfitting model with the training data.

This problem can be reduced by either using a different model, by using better features that can represent the qualities of the dataset or reducing the constraints on the data. Because this model is not regularized, the last option cannot apply for this model. 

First, we can use a more complex model to see whether that will fix the problem. Let's use Decision Tree regression algorithm. We can train and evaluate the algorithm with the same dataset.


```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(housing_prepared, housing_labels)
```

```python
housing_predictions = tree_reg.predict(housing_prepared)
tree_mse = mean_squared_error(housing_labels, housing_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

The RMSE error of 0.0 shows that the model is completely accurate which is called overfitting. To confirm that the model is overfitted, we can use a different set of data to train and test the model. This could provide more information about the overfitting of the model. 

We should not touch the testing set until we have a model that we are confidant about. k-fold cross validation is a better alternative to use a different set of training and testing data for model evaluation on overfitting. The following code will divide the data set to 10 subsets and then it will train and test the model iteratively 10 time. Each subset will be selected as the testing set and the rest 9 subsets will be collectively used as the training set at the each step.  

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(tree_rmse_scores)
```

After the cross validation it seems like Decision Tree performs worse than the Linear Regression.The cross validation provides the measure of performance and also provides how precise the estimation is with the standard deviation.

Let's compute the same score for the Linear Regresion with cross validation.

```python
lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels,
 scoring="neg_mean_squared_error", cv=10)

lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

Now it is obvious that the Decision Tree regressor overfits badly and it performs worse than the Linear Regression model.

Let's train a Random Forest algorithm with the data. Random Forest is an Ensemble Learning model. The algorithm works by training multiple Decision Trees on randomly selected subsets of the features and averaging the results of the trees. 

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(housing_prepared, housing_labels)
housing_predictions = forest_reg.predict(housing_prepared)
forest_mse = mean_squared_error(housing_labels, housing_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
forest_scores = cross_val_score(forest_reg, housing_prepared, housing_labels, scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```

This is a much better result compared to Decision Tree regressor. The RMSE value is much lower on the training set than the cross validation set, which implies overfitting with the training set. The accuracy of the model can be improved further by simplifying the model, constraining, increasing number of training data or by using different other machine learning algorithms. The goal should be to select a few number of models which provide promising results.

## Fine-Tune the Model

After shortlisting promising modles, the models should be fine-tuned. Grid Search and Randomized Search are two methods used in fini tuning models. 

In Grid Search, we provide the hyperparameters should be experimented with the model. The following code experiments all the combinations of hyperparameters with the model.

```python
from sklearn.model_selection import GridSearchCV
param_grid = [
{'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
{'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3,
4]},
]
forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
scoring='neg_mean_squared_error',
return_train_score=True)
grid_search.fit(housing_prepared, housing_labels)
```

The Grid Search has explored 12+6= 18 combinations of parameters with the algorithm and performed 18 * 5= 90 rounds of training and testing. The best set of parameters and the best estimator can be obtained using the codes below.

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

The evaluation scores are as below.

```python
cvres = grid_search.cv_results_
```

```python
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

The best solution is obtained by setting max_features and n_esimators to 6 and 30 respectively. The RMSE for this combination is 49,981. Which is slightly better than the score obtained with default hyperparameter values (52,784).

The Grid Search approach consumes a very long time when it comes to a large hyperparameter search space. In this case, Randomized Search is preffered. Instead of trying all the hyperparameter combinations, Grid Search evaluates a given number of randomly selected hyperparameter combinations. 

It is a good idea to run a feature importance to find out the most important features to perform regression analysis accurately. After performing this analysis, we can drop relatively less useful features for the study. Further, adding new features or dropping uninformative features can help to get rid of some specific error introduced by data to the analysis.

```python
feature_importances = grid_search.best_estimator_.feature_importances_
feature_importances
```

```python
extra_attribs = ["rooms_per_hhold", "pop_per_hhold", "bedrooms_per_room"]
cat_encoder = full_pipeline.named_transformers_["cat"]
cat_one_hot_attribs = list(cat_encoder.categories_[0])
attributes = num_attribs + extra_attribs + cat_one_hot_attribs
sorted(zip(feature_importances, attributes), reverse=True)
```

## Evaluating the model on the Test Set

Now we have a model with sufficient performance and it is time to evaluate the final model with the test set.  

```python
final_model = grid_search.best_estimator_
X_test = strat_test_set.drop("median_house_value", axis=1)
y_test = strat_test_set["median_house_value"].copy()
X_test_prepared = full_pipeline.transform(X_test)
final_predictions = final_model.predict(X_test_prepared)
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse) # => evaluates to 47,730.2
```
